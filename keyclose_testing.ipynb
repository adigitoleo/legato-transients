```python
import audacity as au
import numpy as np
import scipy.io.wavfile as wf
import scipy.signal as sig
import matplotlib.pyplot as pl
%matplotlib notebook
```

```python
auf1 = au.Aup('key_only_recordings/calibration/1V_test.aup')
auf2 = au.Aup('key_only_recordings/calibration/992mV_test.aup')
auf3 = au.Aup('key_only_recordings/calibration/1.02V_test.aup')
w1 = auf1.get_channel_data(0)
w2 = auf2.get_channel_data(0)
w3 = auf3.get_channel_data(0)
maxargs1 = sig.argrelmax(w1, order=1)[0]
maxargs2 = sig.argrelmax(w2, order=1)[0]
maxargs3 = sig.argrelmax(w3, order=1)[0]
maxvals1 = [ w1[arg] for arg in maxargs1 ]
maxvals2 = [ w2[arg] for arg in maxargs2 ]
maxvals3 = [ w3[arg] for arg in maxargs3 ]
maxvals_all = maxvals1+maxvals2+maxvals3
maxvals23 = maxvals2+maxvals3
```

```python
pl.figure()
pl.plot(maxvals2)
pl.plot(np.arange(len(maxvals2)), np.ones(len(maxvals2))*np.mean(maxvals2))
(ar,br) = np.polyfit(np.arange(len(maxvals2)),maxvals2,1)
yr = np.polyval([ar,br],np.arange(len(maxvals2)))
pl.plot(np.arange(len(maxvals2)), yr, 'r-', linewidth='1')
```

```python
pl.figure()
pl.plot(maxvals3)
pl.plot(np.arange(len(maxvals3)), np.ones(len(maxvals3))*np.mean(maxvals3))
(ar,br) = np.polyfit(np.arange(len(maxvals3)),maxvals3,1)
yr = np.polyval([ar,br],np.arange(len(maxvals3)))
pl.plot(np.arange(len(maxvals3)), yr, 'r-', linewidth='1')
```
